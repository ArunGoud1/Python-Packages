## Matplotlib Line Plot

- You can use the keyword argument linestyle, or shorter ls, to change the style of the plotted line

```python
x = [3,4,5,6,9,4,3,4]
y = [10,3,4,7,8,5,1,7]

plt.plot(x,y)

print(plt.show)
```

## Matplotlib Bar Plot

- it shows all data in Barcharts modes

```python
x = [3,4,5,6,9,4]
y = [10,7,8,5,9,2]

plt.bar(x,y)

```

## Matplotlib Histograms Plot

```python
y = [1,2,3,4,5,6,7]
plt.hist(y)
plt.show()
```

## Matplotlib Scatter Plot

```python
x = [3,4,5,6,9,4,3,4]
y = [10,3,4,7,8,5,1,7]

plt.scatter(x,y)

print(plt.show)
```

##  Matpolotlib Charts

```python

labels = ['Apple','samsung','vivo','onplus']
sizes = [23,35,12,33]
plt.pie(sizes,labels=labels, autopct = '%1.2f%%')
plt.title('Mobile companies Chart')
plt.show()
```

```python
x = [1,2,3,4,5]
y1,y2 = [10,13,15,11,18],[5,7,9,10,11]

plt.fill_between(x, y1,y2, color = 'green', alpha = 0.6 ,label = 'area 1-2')
plt.plot(x, y1, label = 'ameerpet', marker = '*') , 
plt.plot(x, y2, label = 'MGBS', marker = 'x')

plt.xlabel('Ameerpet'), plt.ylabel('mgbs'),plt.title('Area Chart Example')

plt.legend(),plt.show()
```

## Introduction to Figurs

```python
import matplotlib.pyplot as plt

from matplotlib import pyplot as plt

fig= plt.figure(figsize=(5,5))
fig.suptitle('Figure')
ax = fig.add_subplot()
ax.set_title('Axis x,y',loc = 'left',fontstyle = 'oblique', fontsize = 'medium')
```

## Subplot() function

```python
fig = plt.figure(figure = (5,5))
ax1 = fig.add_subplot(111)
ax1.plot([1,2,3])
ax2 = fig.add_subplot(221,facecolor = 'r')
ax2.plot([1,2,3])
```

```python
import numpy as np
```

```python
fig,a = plt.subplots(3,3)
x= np.arange(1,5)
a[0][0].plot(x,x*x)
a[0][0].set_title('square')
a[0][1].plot(x,np.sqrt(x))
a[0][1].set_title('square root')
a[1][0].plot(x,np.exp(x))
a[1][0].set_title('exp')
a[1][1].plot(x,np.log10(x))
a[1][1].set_title('log')
plt.show()
```

```python
fig,a = plt.subplots(10,6)
x= np.arange(1,5)
a[0][0].plot(x,x*x)
a[0][0].set_title('square')
a[0][1].plot(x,np.sqrt(x))
a[0][1].set_title('square root')
a[1][0].plot(x,np.exp(x))
a[1][0].set_title('exp')
a[1][1].plot(x,np.log10(x))
a[1][1].set_title('log')
plt.show()
```

```python
import matplotlib.pyplot as plt

import matplotlib.cbook as cbook
import matplotlib.patches as patches

with cbook.get_sample_data('WIN_20240418_14_48_27_Pro.jpg') as image_file:
    image = plt.imread(image_file)

fig, ax = plt.subplots()
im = ax.imshow(image)
patch = patches.Circle((260, 200), radius=200, transform=ax.transData)
im.set_clip_path(patch)

ax.axis('off')
plt.show()
```
